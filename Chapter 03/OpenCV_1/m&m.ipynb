```python
import cv2
import matplotlib.pyplot as plt
path = 'img\img\m.jpg'
img  = cv2.imread(path)
type(img)
```

```python
color_range = cv2.imread('img\img\hsv.png')
plt.figure(figsize=(20,10))
plt.imshow(color_range)
```

```python
rgb_colors = cv2.cvtColor(color_range, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(20,10))
plt.imshow(rgb_colors)
```

```python
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
```

```python
hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
plt.figure(figsize=(12,8))
lower_bound = (80,100,10)
upper_bound = (110,255,255)
mask = cv2.inRange(hsv_img, lower_bound, upper_bound)
masked_img = rgb_img.copy()
masked_img = cv2.bitwise_and(rgb_img,rgb_img, mask=mask)
plt.title('Blue balls')
plt.imshow(masked_img)
```
