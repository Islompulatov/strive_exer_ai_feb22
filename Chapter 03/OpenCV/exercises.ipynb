```python
text_img = rgb_img.copy()
plt.figure(figsize=(12,8))
cv2.putText(text_img, 'White and distant, absorded in itslef,', (100,50),cv2.FONT_HERSHEY_SCRIPT_SIMPLEX, 1, (255, 0, 0))
cv2.putText(text_img, 'endlessly the sky covers and uncovers,', (80, 370),cv2.FONT_HERSHEY_SCRIPT_SIMPLEX,1, (255,0,0))
cv2.putText(text_img, 'moves and remains.', (80, 400),cv2.FONT_HERSHEY_SCRIPT_SIMPLEX,1, (255,0,0))
plt.imshow(text_img)
```

```python
img2 = cv2.imread('YOLO Resources\img\giraffe.jpg')
rgb_img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(12,8))
plt.imshow(rgb_img2)
```

```python
giraffe_img = rgb_img2.copy()
cv2.rectangle(giraffe_img, (160,400), (350, 50), (200,50,200), 4)
cv2.putText(giraffe_img, 'This is Giraffe', (100,30), cv2.FONT_HERSHEY_TRIPLEX, 1, (255,255,20))
cv2.circle(giraffe_img, (380, 350), 60, (0,255,0), 2)
cv2.putText(giraffe_img, 'This is zebra', (250,450), cv2.FONT_HERSHEY_SIMPLEX, 1, (20,50,250))
plt.figure(figsize=(12,8))
plt.imshow(giraffe_img)
```

```python
R, G, B = cv2.split(giraffe_img)
```

```python
G
```

```python
plt.figure(figsize = (12,8))
fig, axs = plt.subplots(3, sharex='row', sharey='col')
axs[0].imshow(R, cmap = 'gray')
axs[1].imshow(G, cmap = 'gray')
axs[2].imshow(B, cmap = 'gray')
```
