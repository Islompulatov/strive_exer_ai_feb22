```python
import cv2
import matplotlib.pyplot as plt
```

```python
dog = cv2.imread("img\img\dog.jpg")
type(dog)
```

```python
cv2.imshow('dog1', dog)
cv2.waitKey(1)
```

```python
plt.figure(figsize=(12,8))
rgb_dog = cv2.cvtColor(dog, cv2.COLOR_BGR2RGB)
gray_dog = cv2.cvtColor(dog, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_dog, cmap='gray')
```

```python
retval, dst = cv2.threshold(gray_dog, 200, 255, cv2.THRESH_BINARY)
plt.imshow(dst, cmap='gray')
```

```python
retval, dst = cv2.threshold(gray_dog, 200, 255, cv2.THRESH_BINARY_INV)
plt.imshow(dst, cmap='gray')
```

```python
retval, dst = cv2.threshold(gray_dog, 230, 255, cv2.THRESH_TRUNC)
plt.imshow(dst, cmap='gray')
```

```python
retval, dst = cv2.threshold(gray_dog, 120, 255, cv2.THRESH_TOZERO)
plt.imshow(dst, cmap='gray')
```

OTSU THRESHOLD

```python
beach = cv2.imread('img/img/beach.jpg')
type(beach)
```

```python
plt.figure(figsize=(12,8))
gray_beach = cv2.cvtColor(beach, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_beach, cmap='gray')
```

```python
ret, otsu = cv2.threshold(gray_beach, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
```

```python
ret
```

```python
plt.imshow(otsu,cmap='gray')
```

```python
ret, otsu = cv2.threshold(gray_beach, 100, 255, cv2.THRESH_BINARY)
plt.imshow(otsu,cmap='gray')
```

ADAPTIVE THRESHOLD

```python
text_img = cv2.imread('img/img/notes.png')
text_img_gray = cv2.imread('img/img/notes.png', 0)
plt.figure(figsize=(20,10))
plt.imshow(text_img_gray, cmap='gray')
```

```python
dat = cv2.adaptiveThreshold(text_img_gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 3)
plt.figure(figsize=(20,10))
plt.imshow(dat, cmap='gray')
```
