```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Exercise 2: Get the Total profit of all months and show a line plot with the following Style properties:
- Line Style dotted and Line color should be red
- Show legend at the lower right location.
- X label name = Month Number
- Y label name = Sold units number
- Add a circle marker.
- Line marker color as read
- Line width should be 3

The line plot graph should look like this:


```python
plt.figure(figsize=(10,8))
plt.plot(df4['month_number'], df4['total_profit'],linestyle = 'dashed', color = 'r', 
marker = 'o', mfc= 'black', linewidth = '3', label = 'total_profit',)
plt.xlabel('Month Number')
plt.ylabel('Sold units number')
plt.title('Company sales data of last year')
plt.legend(loc = 'lower right')
```

###Â Exercise 5: Read the face cream and facewash product sales data and show it using the bar chart.
The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

The bar chart should look like this:


```python
fig, ax = plt.subplots(figsize = (10,8))
ax.bar([x-0.25 for x in df4['month_number'].tolist()],  df4['facecream'].tolist(), label = 'Faceream sales',width= 0.25, align='edge')
ax.bar([x+0.25 for x in df4['month_number'].tolist()],   df4['facewash'].tolist(), label = 'Facewash sales',width = -0.25, align='edge')
plt.xticks(df4['month_number'].tolist())
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.title('Facewash and Facecream Sales Data')
plt.legend(loc = 'upper left')
plt.grid(True, color = 'grey', linewidth = 2, linestyle ='--')

plt.show()
```

### Exercise 7: Read the total profit of each month and show it using the histogram to see the most common profit ranges.
The histogram should look like this:


```python
plt.figure(figsize=(10,8))
ranges = [150000, 175000, 200000, 225000, 250000, 300000, 350000]
plt.hist(df4['total_profit'], ranges)
plt.xlabel('profit range in dollars')
plt.ylabel('Actual profit in dollar')
plt.title('Profit Data')
plt.show()
```

### Exercise 8: Calculate the total sale data for last year for each product and show it using a Pie chart.
Note: In the Pie chart you should display the Number of units sold per year for each product in percentage.

The Pie chart should look like this:


```python
# Write your code here
charts = ['FaceCream', 'FaseWash', 'ToothPaste', 'Bathing soap', 'Shampoo', 'Moisturizer']
datas =  [df4['facecream'].sum(), df4['facewash'].sum(), df4['toothpaste'].sum(), 
         df4['bathingsoap'].sum(), df4['shampoo'].sum(), df4['moisturizer'].sum()]
plt.pie(datas, labels=charts, autopct='%1.1f%%') 
plt.title('Sales Data')
plt.legend(loc= 'lower right')   
plt.show()     
```

###Â Exercise 9: Read the Bathing soap facewash of all months and display it using the Subplot.
The Subplot should look like this:


```python

fig, ax = plt.subplots(2, sharex=True, figsize = (12,8))
ax[0].plot(df4['month_number'], df4['bathingsoap'], label = 'Bathingsoap Sales Data', color='k', marker='o', linewidth=3)
ax[0].set_title('Sales data of  a Bathingsoap')
ax[1].plot(df4['month_number'], df4['facewash'], label = 'Face Wash Sales Data', color='r', marker='o', linewidth=3)
ax[1].set_title('Sales data of  a facewash')

plt.xticks(df4['month_number'])
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.show()
```
